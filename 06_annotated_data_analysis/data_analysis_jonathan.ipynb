```python
import pandas as pd
import numpy as np
import json

input_path = "/Users/jhoff/Universität St.Gallen/STUD-Capstoneproject Tell 6 - Dokumente/General/02-Coding/01-Data/12_annotated_batches/batch1_jona.jsonl"
```

```python
#data = pd.read_json("/Users/jonathanebner/Universität St.Gallen/STUD-Capstoneproject Tell 6 - General/02-Coding/01-Data/12_annotated_batches/batch1_jonathan.jsonl")
```

```python
with open(input_path, "r", encoding="utf8") as json_file:
    annotated_recipes = [json.loads(line) for line in json_file]
```

```python
data = pd.DataFrame(annotated_recipes)
print("Rows:", data.shape[0])
print("Columns:", data.shape[1])
data.head(2)
```

```python
data2 = pd.DataFrame()
data2["text"] = data["text"]
data2["Anzahl Wörter"] = data["text"].apply(lambda x: len(x.split()))
data2["Anzahl Spans"] = data["spans"].apply(lambda x: len(x))
data2["Anzahl Tokens"] = data["tokens"].apply(lambda x: len(x))
data2["Anzahl Relations"] = data["relations"].apply(lambda x: len(x))
```

```python
data2.head(2)
```

```python
data2["Anzahl Arg0"] = data["relations"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("ARG0"))
data2["Anzahl Arg1"] = data["relations"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("ARG1"))
data2["Anzahl Arg"] = data["relations"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("ARG"))
```

```python
data2.head(2)
```

```python
data2["Anzahl Zutaten"] = data["spans"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("Z"))
data2["Anzahl V"] = data["spans"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("V"))
data2["Anzahl Tools"] = data["spans"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("TOOL"))
data2["Anzahl Zeitpunkte"] = data["spans"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("ZEITP"))
data2["Anzahl Dauer"] = data["spans"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("DAUER"))
data2["Anzahl Attribut"] = data["spans"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("ATTR"))
data2["Anzahl Temperatur"] = data["spans"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("TEMP"))
data2["Anzahl Präposition"] = data["spans"].apply(lambda x: [x[i]["label"] for i in range(len(x))].count("PRÄP"))
```
